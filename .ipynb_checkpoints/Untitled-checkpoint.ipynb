```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filepath = "/Users/alexanderigel/Documents/Data Science Master/Agile Data Science/Project/3102154/"
df = pd.read_csv(filepath + 'amino_acid_genotypes_to_brightness.tsv', sep = '\t', engine = 'python')
df.head()
```

```python
muties = df['aaMutations'].str.split(':').apply(pd.Series)  #returns a dataframe with the aminoacids nicely separated 
muties.head()                                                  # ...and full of NaN's :_(
```

```python
df_sep = pd.concat([muties, df], axis=1)
df_sep = df_sep.drop(['aaMutations'], axis = 1) #droping the 'aaMutations' column 
df_sep.head()
```

```python
mutants = df['aaMutations']       #split to list of aminoacids
mutants = mutants.str.split(':')
mutants.tail()
```

```python
mut_len = np.zeros(len(mutants))  #start from index 1 to drop the first NaN
for i in xrange(1,mutants.shape[0]):
    mut_len[i]=len(mutants[i])
```

```python
df.head(5)
```

```python
muties = pd.concat((muties,df['medianBrightness']), axis=1)
muties
```

```python
singleDf = muties.ix[mut_len==1].dropna(axis=1)
doubleDf = muties.ix[mut_len==2].dropna(axis=1)
tripleDf = muties.ix[mut_len==3].dropna(axis=1)
```

```python
singleDf.head()
```

```python
doubleDf.head()
```

```python
tripleDf.head()
```

```python
singleDf.values[0,0]
```

```python
first_column_app=[]
for i in xrange(len(singleDf)):
    first_column_app.append(sum(doubleDf[0]== singleDf.values[i,0]))
```

```python
singleDf['app_double0'] = first_column_app
```

```python
sec_column_app=[]
for i in xrange(len(singleDf)):
    sec_column_app.append(sum(doubleDf[1]== singleDf.values[i,0]))
```

```python
singleDf['app_double1'] = sec_column_app
```

```python
singleDf.head()
```

```python
singleDf['app_total'] = singleDf.app_double0 + singleDf.app_double1
```
